# 16- splitting a string into multiple columns

```python
# import libraries
import pandas as pd
df = pd.DataFrame({'name': ['afaq ahmad', 'shaheer ali ','farooq shah','khezer khan'],
                   'country':['usa, kpk ','russia, punjab','Londan, sindh','china, Balochistan']})
df
                   
```

```python
df.name.str.split( expand=True)
df
```

```python
# splittig into two columns
df.name.str.split().head()
df
```

```python
# adding  those splits into new columns 
df[["first_name","last_name"]] = df.name.str.split( expand=True)
df.head()
```

```python
import pandas as pd
df[["country","city"]] = df.country.str.split(',', expand=True)
df.head()
```

```python
# refine data manipulation
df = df[['first_name','last_name', 'country','city']]
df
```

# 3- how to rename columns ?

```python
df = pd.DataFrame({'col A':[1,2,3,22,23,24,], 'col B': [4,5,6,7,8,9,]})
df.head()
```

```python
df.rename(columns={'A col':  'col_a','col B': 'col_b'}, inplace=True)
df
```

```python
# rename columns
df.columns=['col_aa','col_bb']
df
```

```python
# to replace any character, string
df.columns=df.columns.str.replace('_','')
df
```

```python
df.columns=df.columns.str.replace('','_')
df.head()
```

```python
# adding prefix to columns
df=df.add_prefix('baba_')
df

```

```python
# Adding suffix to columns
df=df.add_suffix('_baba')
df
```

```python
df.columns=['col_a' , 'col_b'] 
df
```

```python
df=df.add_suffix('_baba')
df
```
